```python
stock_data = data[data["stock_id"] == 1]
stock_data.nunique()
#26455 observations for the first stock id
```

```python
#Check correlation matrix for numeric variables
numeric_d = stock_data.select_dtypes(["number"])
numeric_d = numeric_d.drop(["stock_id","date_id"],axis=1)
n_corr = numeric_d.corr(method='pearson', min_periods=1, numeric_only=False)
```

```python
n_corr
```

```python
sns.heatmap(n_corr, cmap= "Blues")
#Some high correlations are to be espected due to some variables being derived from others
#reference price, bid price, ask size and wap have high corelation with multiple variables. To prevent multicolinearity we will remove the
```

```python
#clean data
#remove 3 variables highly correlated with each other and correlated with "wap", "wap" is kept as its a transformation of the previous variables
#and it has a higher correlation with the target variable
#"far price" will also be removed as it correlates highly with "near price"
data_clean = stock_data.drop(["reference_price","bid_price","ask_price","far_price"], axis=1)
#remove the few missing values
data_clean = data_clean.dropna()
```
